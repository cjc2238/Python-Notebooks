## Reading data using pandas

```python
# conventional way to import pandas
import pandas as pd
```

```python
# read data into a DataFrame
data = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv', index_col=0)
data.head()
```

```python
# display the last 5 rows
data.tail()
```

```python
# check the shape of the dataframe
data.shape
```

What are the **features?**
TV: advertising dollars spent on TV for a single product in a given market (in thousands of dollars)
Radio: advertising dollars spent on Radio
Newspaper: advertising dollars spent on Newspaper
What is the response?
Sales: sales of a single product in a given market (in thousands of widgets)

## Visualizing data using seaborn

```python
# conventionalw ay to import seaborn
import seaborn as sns

# allow plots to appear within the notebook
%matplotlib inline
```

```python
# visualize the relationship between the features and the response variables using scatterplots

sns.pairplot(data, x_vars=['TV','Radio','Newspaper'], y_vars='Sales', size =7, aspect=0.7)
```

```python
# visualize the relationship between the features and the response variables using scatterplots and line of best fit (95% conf.) relaionship

sns.pairplot(data, x_vars=['TV','Radio','Newspaper'], y_vars='Sales', size =7, aspect=0.7, kind='reg')
```

```python
# select a subset of the origional data frame containing all the training features
feature_cols = ['TV','Radio','Newspaper']
X = data[['TV','Radio','Newspaper']]

# print the first 5 rows

X.head()
```

```python
# check the shape of the data grame

print (type(X))
print (X.shape)
```

```python
# select a series from the data frame

y = data['Sales']

y.head()
```

## Split X and y into a training and test set

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# default split is 75% to training and 25% to testing

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Linear Regression in scikit-learn

```python
# import the model
from sklearn.linear_model import LinearRegression

# instantiate
linreg = LinearRegression()

# fit the model with the training data (this learns the coefficients)
linreg.fit(X_train, y_train)
```

# Interpreting the model coefficients



```python
# print the intercept and coefficients

print (linreg.intercept_)
print (linreg.coef_)
```

```python
# pair the feature names with the coefficients

zip(feature_cols, linreg.coef_)
```

## Make predictions


```python
# make predictions on the testing set
y_pred = linreg.predict(X_test)
```

## Evaluate the regression model

```python
# Use Mean Absolute Error, Mean Squared Error, or Root Mean Squared Error{This is best} as a form of model evvaluation for regression metrics
import numpy as np
from sklearn import metrics

print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

# The lower the RMSE the better
```
